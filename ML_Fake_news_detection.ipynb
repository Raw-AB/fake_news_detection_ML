```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_learning_curves
```

```python
df =  pd.read_csv("OnlineNewsPopularity.csv")
df.head()
```

```python
df.drop(labels=['url', ' timedelta'], axis = 1, inplace=True)
df
```

Shares Rating
----------

```python
shares = df[' shares']
df[' shares'].describe()
```

```python
#Classifying the shares
rate = []
for i in shares:
    if i <= 645:
        rate.append(1)
    elif i > 645 and i <= 861:
        rate.append(2)
    elif i > 861 and i <= 1400:
        rate.append(3)
    elif i > 1400 and i <= 31300:
        rate.append(4)
    elif i > 31300 and i <= 53700:
        rate.append(5)
    elif i > 53700 and i <= 77200:
        rate.append(6)
    else:
        rate.append(7)


df = pd.concat([df, pd.DataFrame(rate, columns=['rateOfShares'])], axis=1)
df

```

```python
#showing the classes over the share count
plt.figure(figsize=(10, 10))
plot = sns.countplot(x='rateOfShares',data=df)
```

First :Data visualization.
---------------
A)The most published day:
-------------

```python
#test which the most day that the news published in
weeks_days=df[[' weekday_is_monday',' weekday_is_tuesday',' weekday_is_wednesday', 
                      ' weekday_is_thursday', ' weekday_is_friday',' weekday_is_saturday' ,' weekday_is_sunday' ]]

monday=[]
tuesday=[]
wednesday=[]
thursday=[]
friday=[]
saturday=[]
sunday=[]
for i in list(range(weeks_days.shape[0])):
    for j in list(range(weeks_days.shape[1])):
        if ((j==0) and (weeks_days.iloc[i,j])==1):
            monday.append('Monday')
        elif ((j==1) and (weeks_days.iloc[i,j])==1):
            tuesday.append('Tueday')
        elif ((j==2) and (weeks_days.iloc[i,j])==1):
            wednesday.append('Wednesday')
        elif ((j==3) and (weeks_days.iloc[i,j])==1):
            thursday.append('Thursday')
        elif ((j==4) and (weeks_days.iloc[i,j])==1):
            friday.append('Friday')
        elif ((j==5) and (weeks_days.iloc[i,j])==1):
            saturday.append('Saturday') 
        elif ((j==6) and (weeks_days.iloc[i,j])==1):
            sunday.append('Sunday')
days=[len(monday),len(tuesday),len(wednesday),len(thursday),len(friday),len(saturday),len(sunday)]
days
```

```python
fig = plt.figure(figsize = (13,5))
plt.title("news over different day of week", fontsize = 16)
plt.bar(range(7),days)
plt.ylabel("Count", fontsize = 15)
plt.xlabel("Days", fontsize = 15)
plt.xticks(range(7),['monday','tuesday','wednesday','thursday','friday','saturday','sunday'])
plt.show()
# the most published day is wednesday
```

```python
#hypothesis (selected features )
hypo =df[[' n_tokens_title',' n_tokens_content',' num_hrefs',' num_keywords',
                       ' num_imgs',' num_videos',
                    ' weekday_is_monday',' weekday_is_tuesday',' weekday_is_wednesday', 
                      ' weekday_is_thursday', ' weekday_is_friday',' weekday_is_saturday' ,' weekday_is_sunday', 
          ' data_channel_is_lifestyle',' data_channel_is_entertainment' ,' data_channel_is_bus',
                        ' data_channel_is_socmed' ,' data_channel_is_tech',' data_channel_is_world']]
hypo
```

```python

from sklearn.model_selection import train_test_split

#hypothesis
X=hypo
y=df['rateOfShares'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train= scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Third: ML models.
------------
A) SVM:
-----------------------------

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='linear')
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)
```

```python
svclassifier = SVC(kernel='rbf')
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
accuracy_score(y_pred, y_test)
```

```python
svclassifier = SVC(kernel='poly')
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
accuracy_score(y_pred, y_test)
```
